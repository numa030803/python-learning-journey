```python
# using arguments/parameters
def do_more_things(a, b):
    print("A IS", a, "B IS", b)

do_more_things("hello", 1) # while calling the function we passed arguments!
```

```python
def Numu_baby_girl(A, B, C):
    print("I am", A, ", he is", B, "and that's our baby", C)
Numu_baby_girl("Numa", "Raj", "Aceuu")    
```

```python
# ARGUMENTS BEHAVE LIKE LOCAL VARIABLES INSIDE THAT FUNCTION!
i = 5
def Numa(a):
    a = 10
    print(a) # prints 10 as assigned in the function
Numa(i) # function call!!!!
print(i) # prints 5 as it is outside the function
    
```

```python
# this one is like your scripts with argv
def print_two(*args):
    arg1, arg2 = args
    print(f"arg1: {arg1}, arg2: {arg2}")

# ok, that *args is actually pointless, we can just do this
def print_two_again(arg1, arg2):
    print(f"arg1: {arg1}, arg2: {arg2}")

# this just takes one argument 
def print_one(arg1):
    print(f"arg1: {arg1}")

# this one takes no arguments
def print_none():
    print("I got nothin'.")

print_two("Zed","Shaw")
print_two_again("Zed","Shaw")
print_one("First!")
print_none()
    
```
