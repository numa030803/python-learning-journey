```python
# count even numbers
nums = [1, 2, 2, 3, 2, 5]
count = 0
for num in nums:
    if num % 2 == 0:
        count += 1
        
print(count)
```

```python
# Find the smallest number.
nums = [1, 2, 3, 4, 5]
smallest = nums[0]
for num in nums:
    if smallest > num:
        smallest = num
print(smallest)        
    
```
