```python
# Valid Anagram
# Q) Create a dictinary for, s = "rat"
s = "rat"
seen = {} # initialized dictionary
# count = 0, we do not need a count variable because every character has it's own count and that's why we use a dictionary
for i in range (len(s)):
    ch = s[i] # Go to index i in the string s and store that character in the variable ch
    if ch in seen:
        seen[ch] += 1
    else:
        seen[ch] = 1

print(seen)        
        
        
    
    
```

```python
# Count how many a's are there in banana
s = "banana"
count = 0
for ch in s:
    if ch == 'a':
        count += 1

print(count)        
```
