```python
# Arrays & Hashing
# Check if any number repeats?, here we used Brute Force i.e comparing each pair listed in the array, time complexity O(n^2)
nums = [1, 2, 2, 3, 4]
repeat = False
for i in range (len(nums)):
    for j in range (i+1, len(nums)):
        if nums[i] == nums[j]:
           repeat = True
           break

        if repeat:
           break
            
print(repeat)        
        
        
        
     
        
        
```

```python
# set in Python
# Instead of Brute Force we now use set() to solve this same problem, Check if any number repeats?
nums = [1, 2, 2, 3, 4]
seen = set() # We create an empty set
for num in nums: # We loop through each number in the list one by one.
# If I haven't really added anything in seen as it is an empty set, 
# how are we already checking if num in seen? --> Answer: while we loop through nums, we are performing seen.add() automatically!
    if num in seen:
        print(True)
        break
    seen.add(num)   
else:
        print(False)
```
